```python
driver.get("http://gra108.aca.ntu.edu.tw/regbchk/stu_query.asp")
```

```python
s1 = Select(driver.find_element_by_name("DEP"))
```

```python
departments = []
for option in s1.options:
    text = option.text
    departments.append(text)
    
print(departments[:5])
```
